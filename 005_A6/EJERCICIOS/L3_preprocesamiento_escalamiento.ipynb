## L3 - Preprocesamiento y escalamiento de datos

Notebook generado a partir del ejercicio de aplicación.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.impute import SimpleImputer

# Datos
data = {
    'ID': [1, 2, 3, 4],
    'Edad': [25, 45, 30, 40],
    'Ciudad': ['Madrid', 'Sevilla', 'Madrid', 'Barcelona'],
    'Salario': [30000, 50000, np.nan, 40000]
}

df = pd.DataFrame(data)
print("Datos originales:", df)

# 1. Imputar salario con la media
imputer = SimpleImputer(strategy='mean')
df['Salario'] = imputer.fit_transform(df[['Salario']])
print("Datos con salario imputado:", df)

# 2. Label Encoding
le = LabelEncoder()
df['Ciudad_Label'] = le.fit_transform(df['Ciudad'])
print("Label Encoding:", df[['Ciudad', 'Ciudad_Label']])

# 3. One-Hot Encoding (scikit-learn >= 1.2)
ohe = OneHotEncoder(sparse_output=False)
ciudad_encoded = ohe.fit_transform(df[['Ciudad']])
df_ohe = pd.DataFrame(ciudad_encoded, columns=ohe.get_feature_names_out(['Ciudad']))
print("One-Hot Encoding:", df_ohe)

# 4. Variables Dummy
df_dummy = pd.get_dummies(df['Ciudad'], drop_first=True)
print("Variables Dummy:", df_dummy)

# 5. Min-Max Scaling
scaler_minmax = MinMaxScaler()
df[['Edad_MinMax', 'Salario_MinMax']] = scaler_minmax.fit_transform(df[['Edad', 'Salario']])
print("Min-Max Scaling:", df[['Edad_MinMax', 'Salario_MinMax']])

# 6. Estandarización (Z-Score)
scaler_std = StandardScaler()
df[['Edad_Std', 'Salario_Std']] = scaler_std.fit_transform(df[['Edad', 'Salario']])
print("Estandarización:", df[['Edad_Std', 'Salario_Std']])
```

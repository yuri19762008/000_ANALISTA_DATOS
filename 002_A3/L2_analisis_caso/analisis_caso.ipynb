## 2. IMPLEMENTACIÓN PASO A PASO

### Paso 1: Importación de Librerías Necesarias

```python

import pandas as pd
import numpy as np
from io import StringIO
import warnings
warnings.filterwarnings('ignore')

# Para leer tablas web
from pandas import read_html

print("✓ Librerías importadas correctamente")
```

**Explicación:**
- `pandas (pd)`: Manipulación de datos y DataFrames
- `numpy (np)`: Operaciones numéricas y manejo de valores NaN
- `StringIO`: Crear DataFrames desde cadenas de texto
- `read_html`: Extraer tablas de páginas web
- `warnings`: Suprimir advertencias no críticas


### Paso 2: Carga de Datos desde Diferentes Fuentes

#### 2.1 Cargar Datos desde un Archivo CSV

```python
# Simular creación de archivo CSV con datos
csv_data = """ID,Nombre,Departamento,Salario,Fecha_Ingreso
1,Juan García,Ventas,45000.00,2022-01-15
2,María López,TI,52000.00,2021-06-20
3,Carlos Pérez,Finanzas,48000.00,2022-03-10
4,Ana Martínez,Ventas,,2020-11-05
5,Juan García,Ventas,45000.00,2022-01-15
6,Pedro Rodríguez,TI,51000.00,
7,Laura Gómez,RRHH,46000.00,2023-02-18"""

# Guardar como CSV
with open('empleados.csv', 'w', encoding='utf-8') as f:
    f.write(csv_data)

# Cargar el CSV en un DataFrame
df_csv = pd.read_csv('empleados.csv')
print(" Datos desde CSV cargados:")
print(df_csv)
print(f"\nForma: {df_csv.shape[0]} filas, {df_csv.shape[1]} columnas")
```

#### 2.2 Cargar Datos desde un Archivo Excel

```python
# Crear archivo Excel con datos
excel_data = {
    'ID': [101, 102, 103, 104],
    'Proyecto': ['Proyecto A', 'Proyecto B', 'Proyecto C', 'Proyecto D'],
    'Presupuesto': [50000, 75000, 60000, 80000],
    'Estado': ['Activo', 'Completado', 'Activo', 'En Pausa']
}

df_excel_temp = pd.DataFrame(excel_data)
df_excel_temp.to_excel('proyectos.xlsx', index=False)

# Cargar el archivo Excel
df_excel = pd.read_excel('proyectos.xlsx')
print("\n Datos desde Excel cargados:")
print(df_excel)
print(f"\nForma: {df_excel.shape[0]} filas, {df_excel.shape[1]} columnas")
```

#### 2.3 Extraer Datos desde una Tabla Web (HTML)

```python
# Simular una tabla HTML (en la práctica, usarías una URL real)
html_table = """
<table>
  <thead>
    <tr><th>Ciudad</th><th>Población</th><th>País</th></tr>
  </thead>
  <tbody>
    <tr><td>Santiago</td><td>6224000</td><td>Chile</td></tr>
    <tr><td>Lima</td><td>9747000</td><td>Perú</td></tr>
    <tr><td>Bogotá</td><td>8181000</td><td>Colombia</td></tr>
    <tr><td>Buenos Aires</td><td>2890151</td><td>Argentina</td></tr>
  </tbody>
</table>
"""

# Extraer tabla desde HTML
df_html = pd.read_html(StringIO(html_table))[0]
print("\n📊 Datos desde tabla HTML cargados:")
print(df_html)
print(f"\nForma: {df_html.shape[0]} filas, {df_html.shape[1]} columnas")
```

### Paso 3: Análisis de Calidad de Datos (Antes de Limpiar)

```python
print("\n" + "="*60)
print("ANÁLISIS DE CALIDAD - DATOS ORIGINALES")
print("="*60)

print("\n 1. VALORES NULOS POR COLUMNA:")
print(df_csv.isnull().sum())
print(f"\n 2. PORCENTAJE DE DATOS FALTANTES:")
print((df_csv.isnull().sum() / len(df_csv) * 100).round(2))

print("\n 3. FILAS DUPLICADAS:")
print(f"Total de filas duplicadas: {df_csv.duplicated().sum()}")
print("\nDetalles de duplicados:")
print(df_csv[df_csv.duplicated(subset=['Nombre', 'Departamento'], keep=False)])

print("\n 4. TIPOS DE DATOS:")
print(df_csv.dtypes)

print("\n 5. INFORMACIÓN GENERAL:")
print(df_csv.info())

print("\n 6. ESTADÍSTICAS DESCRIPTIVAS:")
print(df_csv.describe())
```

### Paso 4: Limpieza y Estructuración de Datos

#### 4.1 Eliminar Filas Duplicadas

```python
print("\n" + "="*60)
print("LIMPIEZA DE DATOS - PROCESO DETALLADO")
print("="*60)

# Guardar copia para comparación
df_limpio = df_csv.copy()

print("\n PASO 1: Eliminar duplicados")
print(f"   Antes: {len(df_limpio)} filas")

# Eliminar duplicados exactos basados en columnas específicas
df_limpio = df_limpio.drop_duplicates(subset=['Nombre', 'Departamento'], keep='first')

print(f"   Después: {len(df_limpio)} filas")
print(f"   ✓ {len(df_csv) - len(df_limpio)} filas eliminadas")
```

#### 4.2 Manejo de Valores Nulos

```python
print("\n PASO 2: Manejo de valores nulos")
print(f"   Antes: {df_limpio.isnull().sum().sum()} valores nulos")

# Estrategia 1: Imputar Salario con la media del departamento
print("\n   Estrategia para Salario:")
print("   - Calcular promedio por departamento")

salario_por_depto = df_limpio.groupby('Departamento')['Salario'].mean()
print(f"\n   Salario promedio por departamento:\n{salario_por_depto}")

# Aplicar imputación
df_limpio['Salario'] = df_limpio.groupby('Departamento')['Salario'].transform(
    lambda x: x.fillna(x.mean())
)

# Estrategia 2: Eliminar filas con Fecha_Ingreso faltante
print("\n   Estrategia para Fecha_Ingreso:")
print("   - Eliminar filas con fecha faltante (datos críticos)")

df_limpio = df_limpio.dropna(subset=['Fecha_Ingreso'])

print(f"\n   Después: {df_limpio.isnull().sum().sum()} valores nulos")
print("   ✓ Todos los valores nulos tratados")
```

#### 4.3 Convertir Tipos de Datos

```python
print("\n PASO 3: Conversión de tipos de datos")
print("\n   Tipos antes de conversión:")
print(df_limpio.dtypes)

# Convertir Fecha_Ingreso a datetime
df_limpio['Fecha_Ingreso'] = pd.to_datetime(df_limpio['Fecha_Ingreso'])

# Convertir Salario a tipo numérico (asegurar)
df_limpio['Salario'] = pd.to_numeric(df_limpio['Salario'], errors='coerce')

# ID debe ser entero
df_limpio['ID'] = df_limpio['ID'].astype('int64')

# Categorías para Departamento
df_limpio['Departamento'] = df_limpio['Departamento'].astype('category')

print("\n   Tipos después de conversión:")
print(df_limpio.dtypes)
```

### Paso 5: Transformación y Optimización de Datos

#### 5.1 Seleccionar Columnas Relevantes

```python
print("\n" + "="*60)
print("TRANSFORMACIÓN DE DATOS")
print("="*60)

print("\n PASO 1: Seleccionar columnas relevantes")
print("   Columnas originales:", list(df_limpio.columns))

# Seleccionar solo columnas de interés para el análisis
columnas_relevantes = ['ID', 'Nombre', 'Departamento', 'Salario', 'Fecha_Ingreso']
df_limpio = df_limpio[columnas_relevantes]

print("   Columnas finales:", list(df_limpio.columns))
```

#### 5.2 Renombrar Columnas para Mejorar Legibilidad

```python
print("\n PASO 2: Renombrar columnas para legibilidad")
print("   Nombre antes:", list(df_limpio.columns))

# Crear diccionario de renombrado
renombre_dict = {
    'ID': 'id_empleado',
    'Nombre': 'nombre_completo',
    'Departamento': 'depto',
    'Salario': 'salario_anual',
    'Fecha_Ingreso': 'fecha_ingreso'
}

df_limpio = df_limpio.rename(columns=renombre_dict)

print("   Nombre después:", list(df_limpio.columns))
```

#### 5.3 Ordenar Datos Según Columna Clave

```python
print("\n PASO 3: Ordenar datos por columna clave")
print("   Ordenando por: Salario (descendente)")

df_limpio = df_limpio.sort_values(by='salario_anual', ascending=False).reset_index(drop=True)

print("\n Datos después de ordenar:")
print(df_limpio)
```

### Paso 6: Agregar Análisis Descriptivo

```python
print("\n" + "="*60)
print("ANÁLISIS DESCRIPTIVO")
print("="*60)

print("\n Resumen por Departamento:")
resumen_depto = df_limpio.groupby('depto').agg({
    'id_empleado': 'count',
    'salario_anual': ['mean', 'min', 'max']
}).round(2)

resumen_depto.columns = ['Cantidad', 'Salario_Promedio', 'Salario_Mínimo', 'Salario_Máximo']
print(resumen_depto)

print("\n Información del DataFrame Final:")
print(df_limpio.info())

print("\n Estadísticas Descriptivas:")
print(df_limpio.describe())
```

### Paso 7: Exportación de Datos Procesados

#### 7.1 Exportar a CSV


```python
print("\n" + "="*60)
print("EXPORTACIÓN DE DATOS")
print("="*60)

# Guardar como CSV sin incluir el índice
csv_path = 'empleados_limpiados.csv'
df_limpio.to_csv(csv_path, index=False, encoding='utf-8')

print(f"\n✓ Archivo CSV exportado: {csv_path}")
print(f"  Tamaño: {len(df_limpio)} filas")

# Verificar el contenido
print("\nContenido del CSV exportado:")
df_verificacion = pd.read_csv(csv_path)
print(df_verificacion)
```

#### 7.2 Exportar a Excel con Formato

```python
# Guardar como Excel
excel_path = 'empleados_limpiados.xlsx'
df_limpio.to_excel(excel_path, index=False, sheet_name='Empleados')

print(f"\n Archivo Excel exportado: {excel_path}")
print(f"  Tamaño: {len(df_limpio)} filas")

# Crear Excel con múltiples hojas (análisis adicional)
with pd.ExcelWriter('empleados_analisis_completo.xlsx') as writer:
    # Hoja 1: Datos limpios
    df_limpio.to_excel(writer, sheet_name='Empleados', index=False)
    
    # Hoja 2: Resumen por departamento
    resumen_depto.to_excel(writer, sheet_name='Resumen_Departamento')
    
    # Hoja 3: Estadísticas
    stats = df_limpio[['salario_anual']].describe()
    stats.to_excel(writer, sheet_name='Estadísticas')

print(f"\n Archivo Excel completo exportado: empleados_analisis_completo.xlsx")
print("  Incluye: Datos limpios + Resumen + Estadísticas")
```

## 3. COMPARACIÓN: ANTES Y DESPUÉS

```python
print("\n" + "="*60)
print("COMPARACIÓN ANTES Y DESPUÉS")
print("="*60)

print("\n DATOS ORIGINALES (CSV):")
print(f"  - Filas: {len(df_csv)}")
print(f"  - Valores nulos: {df_csv.isnull().sum().sum()}")
print(f"  - Filas duplicadas: {df_csv.duplicated().sum()}")
print(f"\n{df_csv}")

print("\n DATOS LIMPIOS:")
print(f"  - Filas: {len(df_limpio)}")
print(f"  - Valores nulos: {df_limpio.isnull().sum().sum()}")
print(f"  - Filas duplicadas: {df_limpio.duplicated().sum()}")
print(f"  - Tipos de datos validados: ")
print(f"\n{df_limpio}")

print("\n Calidad de datos mejorada:")
print(f"  - Reducción de filas: {((len(df_csv) - len(df_limpio)) / len(df_csv) * 100):.1f}%")
print(f"  - Completitud de datos: 100%")
print(f"  - Integridad: Validada y optimizada")
```

## 4. RESUMEN DEL PROCESO COMPLETO

```python
print("\n" + "="*60)
print("RESUMEN DEL FLUJO DE TRABAJO")
print("="*60)

flujo = """
┌─────────────────────────────────────────┐
│  1. CARGA DE DATOS                      │
│     ├── CSV (empleados.csv)             │
│     ├── Excel (proyectos.xlsx)          │
│     └── HTML (tabla web)                │
└────────────┬────────────────────────────┘
             │
┌────────────▼────────────────────────────┐
│  2. ANÁLISIS DE CALIDAD                 │
│     ├── Identificar nulos               │
│     ├── Detectar duplicados             │
│     ├── Validar tipos de datos          │
│     └── Revisar estructura              │
└────────────┬────────────────────────────┘
             │
┌────────────▼────────────────────────────┐
│  3. LIMPIEZA DE DATOS                   │
│     ├── Eliminar duplicados             │
│     ├── Imputar o eliminar nulos        │
│     ├── Convertir tipos de datos        │
│     └── Validar restricciones           │
└────────────┬────────────────────────────┘
             │
┌────────────▼────────────────────────────┐
│  4. TRANSFORMACIÓN Y OPTIMIZACIÓN       │
│     ├── Seleccionar columnas            │
│     ├── Renombrar campos                │
│     ├── Ordenar datos                   │
│     └── Crear derivados (si aplica)     │
└────────────┬────────────────────────────┘
             │
┌────────────▼────────────────────────────┐
│  5. EXPORTACIÓN DE DATOS                │
│     ├── CSV (sin índice)                │
│     ├── Excel (múltiples hojas)         │
│     └── Validar integridad              │
└─────────────────────────────────────────┘
"""

print(flujo)
```

## 5. MEJORES PRÁCTICAS IMPLEMENTADAS

| Práctica | Descripción | Beneficio |
|----------|-------------|-----------|
| **Copia de DataFrame** | `df.copy()` antes de modificar | Preservar datos originales |
| **Manejo estratégico de nulos** | Imputación + eliminación selectiva | Optimizar calidad vs cantidad |
| **Conversión de tipos** | Asegurar dtype correcto | Evitar errores en análisis |
| **Eliminación de duplicados** | `drop_duplicates()` con lógica | Garantizar unicidad |
| **Renombrado de columnas** | Nombres descriptivos | Mejorar legibilidad |
| **Ordenamiento** | Datos organizados por clave | Facilitar análisis |
| **Exportación sin índice** | `index=False` | Archivos limpios |
| **Múltiples hojas Excel** | Análisis en contexto | Documentación completa |

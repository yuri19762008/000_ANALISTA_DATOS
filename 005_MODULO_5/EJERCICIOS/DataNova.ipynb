## 1. Planteamiento del problema
La empresa DataNova desea estimar el gasto medio semanal en productos de consumo masivo de los clientes frecuentes de una cadena de supermercados.
​
Se dispone de una muestra aleatoria de 60 clientes y se conoce solo la desviación estándar muestral, no la poblacional.
​

Datos:

    Tamaño de muestra n = 60.
    ​Media muestral x=132.50 USD.
    ​Desviación estándar muestral s=15.40 USD.
    Nivel de confianza: 95%.
​

Objetivo: construir un intervalo de confianza para la media poblacional del gasto semanal y luego repetir el cálculo con 99% de confianza.
​

## 2. Elección de la distribución
Como la desviación estándar poblacional σ es desconocida y se utiliza la desviación estándar muestral s, el enfoque teórico indica usar la distribución t de Student con n−1 grados de libertad.
Sin embargo, para n=60 la t con 59 grados de libertad es muy similar a la normal estándar, por lo que en la práctica ambos enfoques producen resultados casi iguales; aun así, se justifica trabajar con t para ser rigurosos.

## 3. Cálculo del intervalo al 95 %

El intervalo de confianza para la media con varianza poblacional desconocida es:

$$
IC = \bar{x} \pm t_{\alpha/2,\, n-1} \cdot \frac{s}{\sqrt{n}}
$$


    Con n=60 se tienen gl=59 grados de libertad.
    Para un nivel de confianza del 95 %, α=0.05 y se requiere t 0.025,59.

```python
import math
from scipy import stats

# Datos
n = 60
x_bar = 132.50
s = 15.40
confidence_95 = 0.95

alpha_95 = 1 - confidence_95
df = n - 1

t_95 = stats.t.ppf(1 - alpha_95/2, df)
se = s / math.sqrt(n)
margen_95 = t_95 * se

ic_95_inf = x_bar - margen_95
ic_95_sup = x_bar + margen_95

t_95, margen_95, ic_95_inf, ic_95_sup
```


## 4. Interpretación del intervalo al 95 %
El intervalo 
$$
[IC 
95% inf
 ,IC 
95% sup
] 
 $$

 indica el rango de valores plausibles para la media poblacional del gasto semanal de todos los clientes frecuentes.
​
Interpretación en lenguaje natural: si se repitiera el muestreo muchas veces y se construyeran intervalos de confianza de la misma forma, aproximadamente el 95 % de esos intervalos contendrían la verdadera media poblacional.

Para la gerencia, significa que el gasto medio real probablemente se encuentra entre los límites calculados; esto permite:

Definir estrategias de pricing y promociones basadas en el gasto típico.

Estimar ingresos esperados por cliente frecuente.

Segmentar clientes y diseñar programas de fidelización coherentes con su nivel de gasto.

## 5. Cálculo del intervalo al 99 % (Plus)

    Para 99 % de confianza, α=0.01 y se requiere t 0.005,59.


```python
confidence_99 = 0.99
alpha_99 = 1 - confidence_99

t_99 = stats.t.ppf(1 - alpha_99/2, df)
margen_99 = t_99 * se

ic_99_inf = x_bar - margen_99
ic_99_sup = x_bar + margen_99

t_99, margen_99, ic_99_inf, ic_99_sup
```

```python
print("IC 95%: [{:.2f}, {:.2f}]".format(ic_95_inf, ic_95_sup))
print("IC 99%: [{:.2f}, {:.2f}]".format(ic_99_inf, ic_99_sup))
```

El intervalo al 99 % será más ancho que el de 95 % porque se exige mayor nivel de confianza, lo que reduce la precisión (aumenta el margen de error).


## 6. Reflexión para el entregable

A mayor nivel de confianza (99 % vs 95 %), el intervalo se hace más amplio: tenemos más seguridad, pero menos precisión sobre el valor exacto de la media.

En decisiones comerciales, la empresa debe equilibrar precisión y seguridad: un intervalo estrecho ayuda a planificar mejor, pero no debe subestimarse la incertidumbre real de los datos.

Estos intervalos permiten basar decisiones (presupuestos, metas de ventas, diseño de promociones) en evidencia cuantitativa y no solo en intuición.

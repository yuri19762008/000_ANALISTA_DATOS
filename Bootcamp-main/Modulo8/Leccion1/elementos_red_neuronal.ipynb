# Elementos de una red neuronal

En este notebook investigamos y explicamos los elementos básicos de una red neuronal artificial: neuronas, capas, pesos, sesgos, funciones de activación y aplicaciones reales.

## 1. Neuronas artificiales y perceptrón

**Objetivo:** entender qué es una neurona artificial, cómo se relaciona con la neurona biológica y qué es un perceptrón.

### 1.1 Idea general

1. Una neurona artificial recibe varios valores de entrada (features) y produce una salida numérica.
2. Cada entrada tiene un **peso** que indica su importancia.
3. Se suma un **sesgo** (bias) que desplaza la activación.
4. Sobre esa suma se aplica una **función de activación** que define la salida final.

### 1.2 Fórmula del perceptrón

Sea un vector de entradas \(x = (x_1, x_2, \dots, x_n)\) y un vector de pesos \(w = (w_1, w_2, \dots, w_n)\). La neurona calcula primero una combinación lineal:

\[ z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b \] 

Luego aplica una función de activación \(f\):

\[ y = f(z) \] 

En el perceptrón clásico se usa una activación de tipo **escalón** (step): devuelve 1 si \(z > 0\) y 0 si \(z \le 0\).

### 1.3 Mini-ejemplo numérico

Supongamos:

- \(x_1 = 2\), \(x_2 = 1\)
- \(w_1 = 0.8\), \(w_2 = 0.4\)
- \(b = -1\)

1. Calculamos la combinación lineal: \(z = 0.8\cdot 2 + 0.4\cdot 1 - 1 = 1\).
2. Aplicamos la función escalón: como \(z > 0\), la salida es \(y = 1\).
3. Interpretación: la neurona se "activa" (clase positiva).

## 2. Tipos de capas en una red neuronal

**Objetivo:** comprender el rol de la capa de entrada, las capas ocultas y la capa de salida.

### 2.1 Capa de entrada

1. Recibe directamente los datos originales (features).
2. Cada neurona de esta capa representa una característica del dato.
3. Normalmente no aplica activación ni transforma mucho los datos, solo los pasa a la siguiente capa.

### 2.2 Capas ocultas

1. Son las capas intermedias entre la entrada y la salida.
2. Cada neurona realiza la operación \(z = w^T x + b\) seguida de una función de activación no lineal.
3. Permiten que la red aprenda representaciones internas cada vez más abstractas de los datos.
4. Cuantas más capas y neuronas, más complejos pueden ser los patrones que se aprenden (aunque también aumenta el riesgo de sobreajuste).

### 2.3 Capa de salida

1. Toma la representación generada por las capas ocultas y la convierte en la respuesta final.
2. El número de neuronas de salida depende del problema:
   - 1 neurona para regresión o clasificación binaria.
   - \(K\) neuronas para clasificación en \(K\) clases.
3. La función de activación suele ser específica: por ejemplo, sigmoid para probabilidad en binario o softmax para probabilidad en varias clases.

### 2.4 Ejemplo sencillo de arquitectura

Problema con 4 features de entrada y 2 clases de salida:

- Capa de entrada: 4 neuronas (una por feature).
- Capa oculta: 8 neuronas con activación ReLU.
- Capa de salida: 1 neurona con activación sigmoid (probabilidad de clase 1).

## 3. Rol de los pesos y los sesgos (bias)

**Objetivo:** entender cómo los pesos y sesgos determinan lo que la red aprende.

### 3.1 Intuición de los pesos

1. Un peso grande en valor absoluto significa que esa entrada influye mucho en la neurona.
2. Un peso positivo refuerza la entrada: si la entrada aumenta, tiende a aumentar \(z\).
3. Un peso negativo invierte el efecto de la entrada.

### 3.2 Intuición del sesgo

1. El sesgo \(b\) permite desplazar la función de activación hacia la izquierda o derecha.
2. Sin sesgo, la neurona solo podría activar cerca del origen.
3. Con sesgo, la neurona puede activarse incluso si todas las entradas son pequeñas.

### 3.3 Ejemplo numérico comparando sesgos

Neurona con una sola entrada \(x\):

- Caso A: \(w = 1\), \(b = 0\).
- Caso B: \(w = 1\), \(b = -0.5\).

Para \(x = 0.2\):

- Caso A: \(z = 1 \cdot 0.2 + 0 = 0.2\).
- Caso B: \(z = 1 \cdot 0.2 - 0.5 = -0.3\).

Si usamos activación escalón, en el Caso A la neurona se activa (\(z > 0\)) y en el Caso B no. Solo cambiar el sesgo cambia la decisión.

### 3.4 Aprendizaje de pesos y sesgos

1. Durante el entrenamiento se define una **función de pérdida** que mide el error entre las predicciones y los valores verdaderos.
2. Mediante descenso por gradiente y backpropagation se calculan derivadas de la pérdida respecto a cada peso y sesgo.
3. Luego se actualizan los parámetros para reducir la pérdida en cada iteración.

## 4. Funciones de activación

**Objetivo:** comparar al menos tres funciones de activación y ver cuándo se usan.

### 4.1 Sigmoid

1. Definición: \(\sigma(x) = \frac{1}{1 + e^{-x}}\).
2. Rango: valores entre 0 y 1.
3. Interpretación: se puede leer como una probabilidad.
4. Uso típico: neuronas de salida en clasificación binaria.

### 4.2 Tanh

1. Definición: \(\tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}\).
2. Rango: valores entre −1 y 1.
3. Está centrada en 0, lo que puede ayudar al entrenamiento.
4. Uso típico: capas ocultas en redes clásicas.

### 4.3 ReLU (Rectified Linear Unit)

1. Definición: \(\text{ReLU}(x) = \max(0, x)\).
2. Si la entrada es negativa, la salida es 0; si es positiva, la salida es la propia entrada.
3. Uso típico: capas ocultas en redes profundas modernas (CNN, MLP, etc.).
4. Ventaja: es simple de calcular y ayuda a evitar el desvanecimiento del gradiente.

### 4.4 Resumen en tabla

| Función | Fórmula (idea)                       | Rango  | Uso típico                       |
|--------|---------------------------------------|--------|----------------------------------|
| Sigmoid| \(1 / (1 + e^{-x})\)                  | (0, 1) | Probabilidades en binario       |
| Tanh   | \((e^{x} - e^{-x})/(e^{x} + e^{-x})\) | (−1, 1)| Capas ocultas tradicionales     |
| ReLU   | \(\max(0, x)\)                      | [0, ∞) | Capas ocultas en redes profundas |

## 5. Aplicaciones reales de las redes neuronales

**Objetivo:** ver en qué contextos prácticos son especialmente útiles las redes neuronales.

1. Visión por computador y autos autónomos: redes convolucionales (CNN) reconocen peatones, señales de tráfico y carriles a partir de imágenes.
2. Reconocimiento de voz y asistentes virtuales: redes recurrentes y transformadores convierten audio en texto y generan respuestas.
3. Detección de fraude: modelos analizan patrones de transacciones para detectar operaciones inusuales.
4. Medicina: análisis de imágenes médicas (radiografías, resonancias) y datos clínicos para apoyar diagnósticos.

## 6. Ejemplo en código: neurona simple en Python

A continuación creamos una neurona artificial muy sencilla en Python para ver los cálculos de forma práctica.

```python
import math

# Funciones de activación
def step(x):
    return 1 if x > 0 else 0

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def relu(x):
    return max(0, x)

# Neurona genérica
def neurona(x, w, b, activacion="step"):
    """Neurona artificial sencilla.

    x: lista de entradas
    w: lista de pesos
    b: sesgo
    activacion: "step", "sigmoid" o "relu"
    """
    z = sum(x_i * w_i for x_i, w_i in zip(x, w)) + b
    if activacion == "step":
        return step(z)
    elif activacion == "sigmoid":
        return sigmoid(z)
    elif activacion == "relu":
        return relu(z)
    else:
        raise ValueError("Función de activación no soportada")

# Probar la neurona con un ejemplo
entradas = [2, 1]
pesos = [0.8, 0.4]
bias = -1

print("Salida con step:", neurona(entradas, pesos, bias, activacion="step"))
print("Salida con sigmoid:", neurona(entradas, pesos, bias, activacion="sigmoid"))
print("Salida con ReLU:", neurona(entradas, pesos, bias, activacion="relu"))
```

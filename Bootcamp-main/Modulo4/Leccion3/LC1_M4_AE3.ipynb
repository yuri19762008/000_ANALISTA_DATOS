# Live Coding : Visualización e interpretación de una correlación “real”

## ¿Qué es un scatterplot y para qué sirve?

Un scatterplot (gráfico de dispersión) muestra la relación entre dos variables numéricas.

Cada punto representa una observación:

eje X → una variable

eje Y → otra variable

Sirve para:

* detectar si existe relación
* ver si es positiva, negativa o inexistente
* observar patrones, tendencias o anomalías


## Dataset simple: ejercicio semanal vs nivel de energía

Supongamos que preguntamos a varias personas:

cuántas horas hacen ejercicio a la semana

cómo evalúan su nivel de energía diario (escala 1–10)

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 120

horas = np.random.randint(0, 11, n)  # 0 a 10 horas

energia = 3 + 0.7 * horas + np.random.normal(0, 1.5, n)
energia = np.clip(energia, 1, 10)
energia = np.round(energia, 0).astype(int)  # energía como “nota” entera

df = pd.DataFrame({
    "Horas_ejercicio": horas,
    "Nivel_energia": energia
})

df.head()
```

¿Esperan que estas variables estén relacionadas? ¿Por qué?

## Representación visual (scatterplot)

```python
import matplotlib.pyplot as plt

plt.scatter(df["Horas_ejercicio"], df["Nivel_energia"])
plt.xlabel("Horas de ejercicio semanal")
plt.ylabel("Nivel de energía diario")
plt.title("Ejercicio vs nivel de energía")

plt.grid(True)
plt.show()
```

¿Los puntos suben de izquierda a derecha?

¿Hay una tendencia clara?

¿Está todo muy disperso o relativamente alineado?

## Cálculo del coeficiente de correlación de Pearson

calcula el coeficiente de correlación de Pearson

```python
df[["Horas_ejercicio", "Nivel_energia"]].corr()
```

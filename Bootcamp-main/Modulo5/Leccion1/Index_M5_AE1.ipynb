*Este ejercicio no busca demostrar una verdad definitiva, sino mostrar cómo el método científico y el análisis estadístico nos ayudan a formular preguntas, interpretar datos y reconocer los límites de nuestras conclusiones.*

1. Pregunta de investigación:

    **¿Existe una relación entre el nivel educativo y el ingreso promedio?**

2. Hipótesis
    ***A mayor nivel educativo, mayor ingreso promedio mensual.***

3. Variables del estudio
    * Variable explicativa (independiente): Nivel educativo
    * Variable respuesta (dependiente): Ingreso promedio mensual

4. Datos: Dataset público con información anonimizada sobre educación e ingresos

    Los datos:
    * ❌ no fueron recolectados por nosotros
    * ❌ no provienen de un experimento
    * ✔ corresponden a un estudio observacional

5. Análisis exploratorio:
    
    Visualizamos los datos:
   - ingreso por nivel educativo
   - distribución y dispersión de ingresos
   - Observamos patrones iniciales y diferencias aparentes entre grupos

6. Prueba estadística (enfoque conceptual)
    
    Dependiendo del tipo de variables, podríamos aplicar:
    * Comparación de medias (si analizamos grupos educativos)
    * Modelo de regresión simple (si tratamos el nivel educativo como variable ordenada)

7. Sesgos y limitaciones

   Este análisis puede verse afectado por:
   * correlación ≠ causalidad
   * variables omitidas (edad, experiencia, región)
   * sesgo de selección
   * datos autoreportados

8. Conclusiones posibles
   - Con estos datos podemos:
      * observar asociaciones
      * cuantificar relaciones

   - Pero no podemos afirmar causalidad

### “El uso del celular antes de dormir reduce la calidad del sueño en adultos jóvenes.”

1️⃣ ¿Es una afirmación científica?
Sí, puede considerarse una afirmación científica porque:

* Es refutable: se puede imaginar un estudio donde adultos jóvenes que usan celular antes de dormir no presenten peor sueño que quienes no lo usan, lo que refutaría la afirmación.

* Es verificable empíricamente: se puede medir uso de celular (minutos, tipo de contenido, brillo, etc.) y calidad del sueño (horas dormidas, latencia de sueño, despertares, registros ) y comprobar si existe o no diferencia.

---

2️⃣ Tipo de método de investigación más adecuado
El método más adecuado sería cuantitativo, porque:

* Se quiere medir variables numéricas: tiempo de uso de celular, duración del sueño, índice de calidad del sueño, frecuencia de insomnio, etc.

* Interesa establecer relaciones (asociación o posible causalidad) mediante análisis estadístico, idealmente con diseños experimentales o cuasi-experimentales (por ejemplo, un ensayo controlado).

Podrían añadirse elementos cualitativos (por ejemplo, entrevistas sobre la experiencia subjetiva del sueño), pero eso sería complementario; el núcleo para evaluar la afirmación es cuantitativo.

---


3️⃣ Diseño breve de una investigación
Un posible diseño:

Población y muestra

* Población objetivo: adultos jóvenes de 18 a 30 años.

* Muestra: por ejemplo, 120 participantes reclutados en universidades o institutos, sin trastornos graves de sueño diagnosticados.

Tipo de estudio

* Ensayo experimental aleatorizado (ECA) de 4 semanas:

    * Grupo experimental: usa el celular al menos 30 minutos en la hora previa a dormir.

    * Grupo control: no usa celular ni otras pantallas en la hora previa a dormir; puede leer en papel o realizar actividades sin pantallas.

Variables y mediciones

* Variable independiente: condición de uso del celular (sí/no, y además registrar minutos reales de uso con una app de monitoreo).

    * Variables dependientes de sueño:

        * Duración total del sueño por noche (en minutos).

        * Latencia de inicio del sueño (minutos en quedarse dormido).

        * Calidad del sueño mediante cuestionarios validados (por ejemplo, PSQI).

        * Somnolencia diurna autorreportada.

    * Instrumentos:

        * Cuestionarios de sueño estandarizados.

        * Actígrafos o relojes inteligentes para registrar patrones de sueño, si el presupuesto lo permite.

* Procedimiento básico

    * Semana 0: línea base. Se registra durante 7 días el patrón de sueño habitual y uso real del celular (sin cambios de conducta).

    * Aleatorización: se asigna al azar a grupo experimental o control.

    * Semanas 1–3:

        * Grupo experimental: se indica usar el celular como mínimo 30 minutos antes de dormir, preferentemente redes sociales o mensajería, manteniendo horaria habitual de acostarse.

        * Grupo control: se prohíbe el uso de pantallas en la última hora; se les pide registrar actividades realizadas en ese período.

    * Se monitorean:

        * Diarios de sueño.

        * Registros de uso del celular (aplicación de seguimiento).

        * Cuestionarios semanales de calidad del sueño y somnolencia diurna.

* Análisis de datos

    * Comparar medias de calidad del sueño y horas dormidas entre grupos al final del estudio, controlando la línea base (ANOVA o modelos lineales).

    * Analizar si mayor tiempo de pantalla se asocia con peor calidad del sueño (correlaciones, regresión lineal o logística para riesgo de insomnio).

* Criterio de interpretación

    * Si el grupo que usa celular antes de dormir muestra significativamente peor calidad de sueño (menos horas, más insomnio, peor puntuación en PSQI) que el grupo que no lo usa, se consideraría que los datos apoyan la afirmación; si no hay diferencias significativas, la afirmación se vería cuestionada.

Ejemplo modelo resuelto

Afirmación: “Las personas que hacen ejercicio regularmente viven más años.”

1️⃣ ¿Puede considerarse una afirmación científica?

Sí, puede considerarse una afirmación científica, pero requiere precisiones.

Es refutable, ya que podría ser contradicha si la evidencia mostrara que no existe diferencia en la esperanza de vida entre personas que hacen ejercicio regularmente y aquellas que no.

Es verificable, en la medida en que tanto el nivel de actividad física como la esperanza de vida pueden definirse y medirse mediante datos observacionales.

Sin embargo, la afirmación es incompleta, ya que no especifica qué se entiende por “hacer ejercicio regularmente” ni bajo qué condiciones se mide “vivir más años”, por lo que necesita ser operacionalizada.

2️⃣ Tipo de método de investigación más adecuado

El enfoque más adecuado sería cuantitativo.

La afirmación propone una relación entre variables medibles y permite ser analizada mediante datos de salud, estudios longitudinales o registros poblacionales.

También podría abordarse con un enfoque mixto si se quisiera complementar el análisis con información cualitativa (por ejemplo, hábitos o percepciones), pero el núcleo del análisis es cuantitativo.

3️⃣ Diseño de investigación (resumen)

Se podría diseñar un estudio observacional longitudinal, utilizando datos provenientes de encuestas de salud o registros poblacionales.

Variables
* Variable explicativa: nivel de actividad física
   * Por ejemplo: frecuencia semanal de ejercicio, minutos de actividad moderada/intensa.
* Variable respuesta: esperanza de vida o mortalidad durante un período de seguimiento.


Análisis
* Clasificación de la población según niveles de actividad física
* Comparación de tasas de mortalidad o esperanza de vida entre grupos
* Modelos estadísticos (por ejemplo, regresión de supervivencia) para controlar variables adicionales

Posibles variables de confusión
* Edad
* Dieta
* Acceso a servicios de salud
* Tabaquismo
* Nivel socioeconómico

4️⃣ Sesgos y limitaciones

* Correlación no implica causalidad: las personas que hacen ejercicio pueden tener, en general, otros hábitos saludables.

* Sesgo de selección: personas más sanas tienden a hacer más ejercicio.

* Datos autoreportados: la actividad física suele ser declarada por los propios participantes.

* Dificultad ética y práctica para realizar experimentos controlados a largo plazo.

**Conclusión del análisis**

Con este diseño, se podría observar y cuantificar una asociación entre actividad física y mayor esperanza de vida.

No obstante, los resultados no permitirían afirmar con certeza que el ejercicio cause directamente una mayor longevidad, aunque sí aportarían evidencia consistente en favor de la relación observada.

## Ejemplo integrado de Inferencia Estadística

**Caso: actividad física y presión arterial**

Contexto:  Queremos estudiar si las personas que hacen ejercicio regularmente presentan menor presión arterial promedio que aquellas que no hacen ejercicio.

No podemos medir a toda la población, así que trabajamos con una muestra.

**1. Estimación de parámetros**

Qué hacemos:

A partir de una muestra de personas adultas, calculamos:
* la presión arterial promedio de quienes hacen ejercicio regularmente
* la presión arterial promedio de quienes no hacen ejercicio

Ejemplo:
* Promedio grupo con ejercicio: 122 mmHg
* Promedio grupo sin ejercicio: 128 mmHg

¿Qué significa esto?:
* Estamos estimando un parámetro poblacional (la presión promedio real de cada grupo) usando datos muestrales.

📌 Nota: Estos valores son estimaciones, no la verdad absoluta de la población.


**2.Intervalos de confianza**

Qué hacemos ahora:

En vez de quedarnos solo con un número, construimos intervalos de confianza para cada promedio.

Ejemplo:
* Grupo con ejercicio:
IC 95%: [120, 124] mmHg

* Grupo sin ejercicio:
IC 95%: [126, 130] mmHg

*Cómo se interpreta*
* Con un 95% de confianza, el valor real poblacional de la presión promedio se encuentra dentro de esos rangos.

📌 Nota: La inferencia estadística reconoce la incertidumbre asociada a trabajar con muestras.

**3.Prueba de hipótesis**

Pregunta formal: ***¿Existe una diferencia estadísticamente significativa en la presión arterial promedio entre personas que hacen ejercicio y las que no?***

Hipótesis

* H₀ (nula): No hay diferencia en la presión arterial promedio entre ambos grupos.
* H₁ (alternativa): Sí existe una diferencia en la presión arterial promedio.


Resultado (supuesto):
* La prueba estadística arroja un p-valor < 0.05.

Qué concluimos
* Rechazamos la hipótesis nula.

La diferencia observada es estadísticamente significativa.

📌 Nota: No estamos diciendo que el ejercicio cause la diferencia, solo que la evidencia apoya una asociación.

**Conclusiones**

Con este análisis podemos:

* ✔ Estimar valores poblacionales a partir de una muestra
* ✔ Cuantificar la incertidumbre de esas estimaciones
* ✔ Evaluar si una diferencia observada es atribuible al azar

📌 Nota:  Pero no podemos afirmar causalidad, ya que:
el estudio es observacional pueden existir variables de confusión (edad, dieta, genética, etc.)

# Live Coding

**Caso: ¿Influye el tipo de música en la concentración?**

**Pregunta de investigación**

¿Influye el tipo de música en la concentración durante una tarea cognitiva?

* Muchas personas estudian escuchando música, pero no está claro si ciertos tipos de música ayudan, perjudican o no tienen efecto sobre la concentración.

**Planteamiento del problema y formulación de hipótesis**

* Hipótesis nula (H₀): El tipo de música no tiene efecto en la concentración.

* Hipótesis alternativa (H₁): El tipo de música sí produce diferencias en el nivel de concentración.


**Identificación de variables**

Variable independiente (VI):

* Tipo de música
* Sin música
* Música clásica
* Música con letra

Variable dependiente (VD):

* Nivel de concentración: Medido como puntaje obtenido en una prueba de atención (por ejemplo, número de aciertos en 10 minutos)

📌 Nota escalas de medición:
* Tipo de música → nominal
* Puntaje de concentración → razón

**Construcción del grupo experimental y grupo control**

Supongamos:

60 participantes en total

Se dividen en 3 grupos de 20 personas cada uno

| Grupo                | Condición        |
| -------------------- | ---------------- |
| Grupo control        | Sin música       |
| Grupo experimental 1 | Música clásica   |
| Grupo experimental 2 | Música con letra |


**Aplicación de aleatorización**

Los participantes se asignan al azar a los tres grupos.

¿Por qué?

* Para reducir sesgo de selección
* Para balancear variables no controladas (edad, habilidad previa, estado de ánimo)

**Generación de datos simulados**

| Participante | Grupo            | Puntaje |
| ------------ | ---------------- | ------- |
| P1           | Sin música       | 62      |
| P2           | Sin música       | 58      |
| P21          | Música clásica   | 71      |
| P22          | Música clásica   | 69      |
| P41          | Música con letra | 55      |
| P42          | Música con letra | 57      |


**Cálculo de media y valor p**

Calculamos la media de concentración por grupo:

| Grupo            | Media |
| ---------------- | ----- |
| Sin música       | 60    |
| Música clásica   | 70    |
| Música con letra | 56    |


Luego aplicamos una prueba estadística:

* ANOVA (si comparas los tres grupos)
* o t-test (si simplificas a dos grupos)

Resultado (ejemplo):

* valor p = 0.02

📌 Nota: p < 0.05 → rechazamos H₀

**Conclusión e interpretación crítica**

*Conclusión estadística*:

Existen diferencias estadísticamente significativas entre los grupos.

*Interpretación crítica:*

* El resultado es estadísticamente significativo ✔
* No implica que el efecto sea grande o relevante para todas las personas ⚠
* El experimento es artificial → validez externa limitada
* No generaliza a todos los contextos (otros tipos de tareas, música, personas)

Cierre:

Podemos concluir que, en este contexto experimental, el tipo de música se asocia a diferencias en concentración, pero no podemos afirmar que esto ocurra siempre o en la vida real.

**Esta demo muestra cómo el método científico y la estadística nos ayudan a pasar de una intuición cotidiana a una conclusión razonada, con datos, control y pensamiento crítico.**

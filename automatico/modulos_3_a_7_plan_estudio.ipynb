### 3.4 Valores perdidos y outliers

Pasos sugeridos:
1. Crear un DataFrame con valores perdidos (`NaN`).
2. Contar y localizar los valores perdidos.
3. Probar eliminación e imputación (media/mediana/moda).
4. Simular outliers y filtrarlos según reglas.


```
import numpy as np

# 1) DataFrame con valores perdidos

df_nan = pd.DataFrame({
    "x": [1, 2, np.nan, 4, 100],  # 100 como posible outlier
    "y": [10, np.nan, 30, 40, 50]
})

# 2) Conteo de valores perdidos
conteo_nan = df_nan.isna().sum()

# 3) Imputación sencilla
media_x = df_nan["x"].mean()
df_imputado = df_nan.copy()
df_imputado["x"] = df_imputado["x"].fillna(media_x)

# 4) Detección simple de outliers (regla 1.5*IQR)
q1 = df_nan["x"].quantile(0.25)
q3 = df_nan["x"].quantile(0.75)
iqr = q3 - q1
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

outliers = df_nan[(df_nan["x"] < lim_inf) | (df_nan["x"] > lim_sup)]

conteo_nan, df_imputado, outliers
```

## Módulo 5 – Inferencia estadística

Objetivo: usar probabilidad, distribuciones, intervalos de confianza y pruebas de hipótesis para inferir sobre la población a partir de una muestra.

### 5.1 Ejemplo de distribución normal y TLC

Pasos sugeridos:
1. Simular una población normal.
2. Tomar muchas muestras y calcular la media de cada una.
3. Observar cómo se distribuyen las medias (TLC).


```
np.random.seed(0)

# 1) Población
poblacion = np.random.normal(loc=50, scale=10, size=100000)

# 2) Muestras y medias
n_muestra = 50
n_iter = 1000
medias = []
for _ in range(n_iter):
    muestra = np.random.choice(poblacion, size=n_muestra, replace=False)
    medias.append(muestra.mean())

medias = np.array(medias)

# 3) Histograma de medias
plt.hist(medias, bins=30)
plt.title("Distribución de medias muestrales (TLC)")
plt.show()

medias.mean(), medias.std()
```

### 5.2 Intervalo de confianza para una media (desviación desconocida)

Pasos sugeridos:
1. Tomar una muestra de la población.
2. Calcular la media muestral y la desviación estándar muestral.
3. Calcular el intervalo de confianza usando t de Student.


```
from scipy import stats

muestra = np.random.choice(poblacion, size=40, replace=False)
media_m = muestra.mean()
std_m = muestra.std(ddof=1)

alpha = 0.05
n = len(muestra)

t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

margen_error = t_crit * std_m / np.sqrt(n)

ic_inf = media_m - margen_error
ic_sup = media_m + margen_error

media_m, (ic_inf, ic_sup)
```

## Módulo 6 – Aprendizaje de máquina supervisado

Objetivo: construir modelos de regresión y clasificación supervisados en Python, controlando sobreajuste mediante validación cruzada.

### 6.1 Flujo básico de un problema supervisado

Pasos típicos:
1. Definir el problema (regresión o clasificación).
2. Separar en train/test.
3. Entrenar un modelo base.
4. Evaluar con métricas adecuadas.
5. Usar validación cruzada para estimar desempeño y evitar sobreajuste.


```
from sklearn.datasets import load_boston, load_iris
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

# Ejemplo de regresión (usamos Boston si está disponible; si no, generamos datos sintéticos)
try:
    datos_reg = load_boston()
    X_reg = datos_reg.data
    y_reg = datos_reg.target
except:
    X_reg = np.random.rand(200, 3)
    y_reg = X_reg @ np.array([3.0, -2.0, 1.0]) + np.random.normal(0, 0.5, size=200)

X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

modelo_reg = LinearRegression()
modelo_reg.fit(X_train, y_train)

pred_test = modelo_reg.predict(X_test)

rmse = mean_squared_error(y_test, pred_test, squared=False)
r2 = r2_score(y_test, pred_test)

# Validación cruzada
cv_scores = cross_val_score(modelo_reg, X_reg, y_reg, cv=5, scoring="r2")

rmse, r2, cv_scores.mean()
```

### 6.2 Ejemplo de clasificación con validación cruzada

Pasos sugeridos:
1. Cargar un dataset de clasificación.
2. Entrenar un modelo (p. ej. logística).
3. Evaluar exactitud en test y con validación cruzada.


```
iris = load_iris()
X_clf = iris.data
y_clf = iris.target

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42, stratify=y_clf)

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_c, y_train_c)

pred_c = log_reg.predict(X_test_c)
acc_test = accuracy_score(y_test_c, pred_c)

cv_scores_clf = cross_val_score(log_reg, X_clf, y_clf, cv=5, scoring="accuracy")

acc_test, cv_scores_clf.mean()
```

### 7.1 Ejemplo de clusterización con k-means

Pasos sugeridos:
1. Generar o cargar datos numéricos.
2. Aplicar k-means para encontrar grupos.
3. Visualizar resultados y comentar.


```
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X_blobs, y_blobs = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(X_blobs)
centros = kmeans.cluster_centers_

plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=labels, cmap="viridis", s=20)
plt.scatter(centros[:, 0], centros[:, 1], c="red", s=100, marker="x")
plt.title("Clusterización k-means")
plt.show()
```
